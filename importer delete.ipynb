```python
while True:
    try:
        # Click the Fetch button
        fetch_btn_path = driver.find_element(By.XPATH, '/html/body/app-root/app-epr/app-pibo-operations/div[1]/div[2]/div/div/div/div/div[1]/div[3]/button')
        fetch_btn_path.click()
        print("Clicked Fetch")

        deleted = 0
        for i in range(50):
            try:
                time.sleep(0.5)
                wait = WebDriverWait(driver, 5)
                binn = wait.until(EC.element_to_be_clickable((By.XPATH, '//table/tbody/tr[1]/td[19]/span')))
                print(f"Found BIN button at row {i}")
                binn.click()

                yes = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[text()="YES"]')))
                yes.click()
                deleted += 1
                print(f"Deleted item {i+1}")
            except:
                print(f"No more deletable entries at iteration {i}")
                break  # Break inner loop if no more deletable items

        if deleted == 0:
            print("No entries deleted, exiting loop.")
            break  # Break outer loop if nothing was deleted (no more records)
        
    except Exception as e:
        print(f"Error occurred: {e}")
        break  # Break if fetch button or any main element not found
```
